## Data Augmentation

 - **Deep Learning Models Are "Data Hungry".**
 - **Small Datasets Lead To Overfitting Most Of The Times.**
 - **So Lets Generate More Data By Augmentating The Training Samples.**

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img_path = "Dataset/AnimalDataset/All_Images/horses/horse-100.jpg"

img = image.load_img(img_path, target_size=(100,100))
```

```python
x = image.img_to_array(img)/255.0
plt.imshow(x)
plt.axis("off")
plt.show()
```

```python
x = x.reshape((1,100,100,3)) # A Batch That Contains Only 1 Image
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode="nearest"
)
```

```python
i = 1
plt.figure(figsize=(15,15))
for batch in datagen.flow(x, batch_size=1):
    plt.subplot(5,5,i)
    plt.imshow(image.img_to_array(batch[0]))
    plt.axis("off")
    if i == 25:
        break
    i+=1

plt.tight_layout()
plt.show()
```
